#### Задание 2

```python
t = np.linspace(start = 0, stop = 10, num = 51)
f = np.cos(t)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(t, 
        f,
        marker = 'o',
        color = 'green',
        alpha = 0.3)

ax.set(title = "График f(t)",
       xlabel = "Значения t",
       ylabel = "Значения f",
       xlim = (0.5, 9.5),
       ylim = (-2.5, 2.5))

plt.show()
```

#### Задание 3

```python
x = np.linspace(start = -3, stop = 3, num = 51)

y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6))
fig.subplots_adjust(left=0.2, wspace=0.3, hspace=0.3)

box = dict(facecolor='yellow', pad=5, alpha=0.2)

ax1 = axs[0, 0]
ax1.set_title('График y1', bbox=box)
ax1.set_xlim(-5, 5)
ax1.plot(x, y1)

ax3 = axs[1, 0]
ax3.set_title('График y3', bbox=box)
ax3.plot(x, y3)

ax2 = axs[0, 1]
ax2.set_title('График y2', bbox=box)
ax2.plot(x, y2)

ax4 = axs[1, 1]
ax4.set_title('График y4', bbox=box)
ax4.plot(x, y4)

plt.show()
```

```python
import pandas as pd
```

```python
DATASET_PATH = 'c:\\Users\\andre\\source\\repos\\python\\DataScience\\creditcard.csv'
df = pd.read_csv(DATASET_PATH)
```

```python
idx = pd.Index(df.Class)
x = np.array([0, 1])
y = idx.value_counts().to_numpy()
```

```python
plt.style.use('fivethirtyeight')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(6, 4))
fig.subplots_adjust(left=0.2, wspace=0.5)
box = dict(facecolor='darkblue', pad=5, alpha=0.2)

ax1.bar(x, 
       y,
       color='cyan')
ax1.set_title('Обычный', bbox=box)

ax2.bar(x, 
       y,
       color='cyan')
ax2.set_yscale('log')
ax2.set_title('Логарифмический', bbox=box)

plt.show()
```
