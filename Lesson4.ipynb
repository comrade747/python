#### Задание 1

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
plt.rcParams['svg.fonttype'] = 'none'
```

```python
x = np.array([1, 2, 3, 4, 5, 6, 7], dtype=int)
y = np.array([3.5, 3.8, 4.2, 4.5, 5, 5.5, 7], dtype=float)
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.scatter(x, y)
plt.show()
```

#### Задание 2

```python
t = np.linspace(start = 0, stop = 10, num = 51)
f = np.cos(t)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(t, 
        f,
        marker = 'o',
        color = 'green',
        alpha = 0.3)

ax.set(title = "График f(t)",
       xlabel = "Значения t",
       ylabel = "Значения f",
       xlim = (0.5, 9.5),
       ylim = (-2.5, 2.5))

plt.show()
```

#### Задание 3

```python
x = np.linspace(start = -3, stop = 3, num = 51)

y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6))
fig.subplots_adjust(left=0.2, wspace=0.3, hspace=0.3)

box = dict(facecolor='yellow', pad=5, alpha=0.2)

ax1 = axs[0, 0]
ax1.set_title('График y1', bbox=box)
ax1.set_xlim(-5, 5)
ax1.plot(x, y1)

ax3 = axs[1, 0]
ax3.set_title('График y3', bbox=box)
ax3.plot(x, y3)

ax2 = axs[0, 1]
ax2.set_title('График y2', bbox=box)
ax2.plot(x, y2)

ax4 = axs[1, 1]
ax4.set_title('График y4', bbox=box)
ax4.plot(x, y4)

plt.show()
```
